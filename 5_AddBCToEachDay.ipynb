#Add average BC for each day

```python
df.head(31)
```

```python
ClangQL_df_cleaned.head(30)
```

```python
def calculate_average_bc(df, stargazers_df):
    average_bc = []
    start_idx = 0

    for i, row in df.iterrows():
        num_stargazers = row['stargazers_count']
        end_idx = start_idx + num_stargazers

        if end_idx <= len(stargazers_df):
            avg_bc = stargazers_df['Betweenness Centrality'].iloc[start_idx:end_idx].mean()
            average_bc.append(avg_bc)
        else:
            average_bc.append(None)

        start_idx = end_idx

    return average_bc
```

```python
repo_name = 'AmrDeveloper/ClangQL'
repo_df = df[df['full_name'] == repo_name].reset_index(drop=True)
repo_df['average_bc'] = calculate_average_bc(repo_df, ClangQL_df_cleaned)

print(repo_df)
```

```python
print(repo_df.isna().sum())
repo_df['average_bc'] = repo_df['average_bc'].fillna(0)
print(repo_df.isna().sum())
```

```python
def process_repo(repo_name, stargazers_df):
    repo_df = df[df['full_name'] == repo_name].reset_index(drop=True)
    repo_df['average_bc'] = calculate_average_bc(repo_df, stargazers_df)
    return repo_df

processed_ore_cli_gpu_df = process_repo('BenjaSOL/ore-cli-gpu', ore_df_cleaned)
print(processed_ore_cli_gpu_df)
```

```python
print(processed_ore_cli_gpu_df.isna().sum())
processed_ore_cli_gpu_df['average_bc'] = processed_ore_cli_gpu_df['average_bc'].fillna(0)
print(processed_ore_cli_gpu_df.isna().sum())
```

```python
processed_nix_inspect_df = process_repo('bluskript/nix-inspect', nix_df_cleaned)
print(processed_nix_inspect_df)
```

```python
print(processed_nix_inspect_df.isna().sum())
processed_nix_inspect_df['average_bc'] = processed_nix_inspect_df['average_bc'].fillna(0)
print(processed_nix_inspect_df.isna().sum())
```

```python
processed_ore_miner_df = process_repo('tonyke-bot/ore-miner', miner_df_cleaned)
print(processed_ore_miner_df)
```

```python
print(processed_ore_miner_df.isna().sum())
processed_ore_miner_df['average_bc'] = processed_ore_miner_df['average_bc'].fillna(0)
print(processed_ore_miner_df.isna().sum())
```

```python
processed_good_training_language_df = process_repo('tsoding/good_training_language', good_df_cleaned)
print(processed_good_training_language_df)
```

```python
final_df = pd.concat([repo_df, processed_ore_cli_gpu_df, processed_nix_inspect_df, processed_ore_miner_df, processed_good_training_language_df], ignore_index=True)
print(final_df)
```
